```python
# 获取页面内容
def getHTMLText(url):
    try:
        r = requests.get(url)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return ""
```

```python
# http://www.gpcxw.com/gongsi/ 爱股网 上市公司资料
# http://www.gpcxw.com/gongsi/600870.html
pattern = re.compile(u'[1-9]\d{3}-\d{2}-\d{2}')
def get_publish_time(code):
    url = "http://www.gpcxw.com/gongsi/" + code + ".html"
    html = getHTMLText(url) # str
#     soup = BeautifulSoup(html,'html.parser') # 解析网页
#     a = soup.find("pre")
    publish_time.append(pattern.findall(html)[1])
```

```python
stock_name = ts.get_hs300s()['name'].values.tolist()

# 比较m_time是否大(晚)于n_time
import time
def compare_time(m_time,n_time):
    m_time1 = time.mktime(time.strptime(m_time,'%Y-%m-%d')) # '2017-09-29'
    n_time1 = time.mktime(time.strptime(n_time,'%Y%m%d')) # '20160401'
    if (float(m_time1) >= float(n_time1)):
        return 1
    else:
        return 0

compare_time('2017-09-29','20160401') # 1
compare_time('2015-09-29','20160401') # 0
```
